## Aufgabe 1.3:

```python
import random
import matplotlib.pyplot as plt


def randSeq(length=500,s=12345):
   random.seed(s)
   return [random.random() for i in range(length)]

def test(n=500, l=5):
   rSeq = randSeq(length = n)
   points = zip(rSeq[0::2],rSeq[l::2])
   plt.figure(figsize=(20,10))
   plt.subplot(2,1,1)
   plt.title("A Random Sequence")
   plt.plot(rSeq)
   plt.subplot(2,1,2)
   plt.title("Scatter Plot")
   plt.scatter(*zip(*points))
   plt.tight_layout()
   plt.show()

test()
```

2. Setze $Y=X^2$

Offensichtlich, $Y$ ist abhängig von $X$.

$EXY=E(X^3)=0$, dann $Cov(X,Y)=EXY-EXEY=0$. So dass $X$ und $Y$ ist unkorreliert.
